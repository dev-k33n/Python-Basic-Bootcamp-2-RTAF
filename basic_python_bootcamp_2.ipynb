```python
# การสร้างตัวแปร

x = 5 # สร้างตัวแปร x จะมีค่าเป็นตัวเลขเท่ากับ 5
y = "Hello" # สร้างตัวแปร x จะมีค่าเป็นตัวอักษรเท่ากับ  Hello

print("x = ", x , "is of type int") #x =  5 is of type int
print("y = ", y , "is now of type str") #y =  Hello is now of type str
```

**Casting**

```python
# การเปลี่ยนชนิดของตัวแปร จาก ตัวเลขเป็น ตัวอักษร หรือ ตัวเลขทศนิยม
x = str(9)    # x จะเป็น '9'
y = int(9)    # y จะเป็น 9
z = float(9)  # z จะเป็น 9.0

print("x = ", x , "is of type ", type(x)) #x =  9 is of type  <class 'str'>
print("y = ", y , "is of type ", type(y)) #y =  9 is of type  <class 'int'>
print("z = ", z , "is of type ", type(z)) #z =  9.0 is of type  <class 'float'>
```

**Variable Names**

```python
# รูปแบบการตั้งตัวแปรที่นิยมและการสร้างตัวแปรให้เข้าใจมากขึ้น

myVariableName = "Hello World Camel Case" # Camel Case

MyVariableName = "Hello World Pascal Case" # Pascal Case

my_variable_name = "Hello World Snake Case" # Snake Case

print("myVariableName : " , myVariableName) #myVariableName :  Hello World Camel Case
print("MyVariableName : " , MyVariableName) #MyVariableName :  Hello World Pascal Case
print("my_variable_name : " , my_variable_name) #my_variable_name :  Hello World Snake Case
```

**Many Values to Multiple Variables**

```python
# การสร้างตัวแปรหลายตัวและกำหนดค่าในตัวแปรหลายตัวแปร

x, y, z = "Apple", "Banana", "Cherry"

print("x : " , x) #x :  Apple
print("y : " , y) #y :  Banana
print("z : " , z) #z :  Cherry
```

```python
# การสร้างตัวแปรหลายตัวและกำหนดค่าในตัวแปรแค่ตัวเดียว
x = y = z = "Orange"

print("x : " , x) #x :  Orange
print("y : " , y) #y :  Orange
print("z : " , z) #z :  Orange
```

**Global Variables**

```python
# สร้างตัวแปรภายนอกฟังก์ชัน และใช้ตัวแปรนั้นภายในฟังก์ชัน

x = "awesome"

def myfunc():
  print("Python is " + x) #Python is awesome

myfunc()
```

```python
# Global Variables
# สร้างตัวแปรภายในฟังก์ชันโดยใช้ชื่อเดียวกับตัวแปรร่วม
x = "awesome"

def myfunc():
  x = "fantastic"
  print("Python is " + x) #Python is fantastic

myfunc()

print("Python is " + x) #Python is awesome
```

```python
# หากต้องการเปลี่ยนค่าของตัวแปรร่วมภายในฟังก์ชัน ให้อ้างอิงถึงตัวแปร

x = "awesome"

def myfunc():
  global x
  x = "fantastic"

myfunc()

print("Python is " + x) #Python is fantastic
```

# **Python Data Types**


data-types-in-python.png

**Built-in Data Types**

*   Text Type:	str
*   Numeric Types:	int, float, complex
*   Sequence Types:	list, tuple, range
*   Mapping Type:	dict
*   Set Types:	set, frozenset
*   Boolean Type:	bool
*   Binary Types:	bytes, bytearray, memoryview
*   None Type:	NoneType

```python
a = "Hello World" 	                                      # *** Data Type 	str ตัวอักษร

b = 20                                                        # *** Data Type 	int ตัวเลขจำนวนเต็ม

c = 20.5                                                      # *** Data Type 	float ตัวเลขทศนิยม

d = 1j                                                         # * Data Type  complex ตัวเลขจำนวนเชิงซ้อน

e = ["apple", "banana", "cherry"]                   # *** Data Type 	list ข้อมูลใน List นั้นจะมีรูปแบบที่คล้ายกับ Array เลยนั่นก็คือ ข้อมูลตัวแรกสุดจะมี Index เป็น 0 และตัวถัด ๆ ไปก็ค่อย ๆ +1 เพิ่มไป

f = ("apple", "banana", "cherry")                    # ** Data Type tuple ชุดของข้อมูล ที่จะแสดงตามลำดับ และไม่สามารถเปลี่ยนแปลงข้อมูลในชุดนั้นได้

g = range(6)                                                # *** Data Type	range เป็นฟังก์ชันที่ทำหน้าที่สร้างลำดับของตัวเลข

h = {"name" : "John", "age" : 36}                  # *** Data Type	dict ชุดข้อมูลแบบ unordered สามารถเปลี่ยนแปลงแก้ไข และสามารถอ้างถึง index ได้

i = {"apple", "banana", "cherry"}                    # * Data Type set ชุดข้อมูลที่มีการแสดงผลไม่เป็นไปตามลำดับ ทำให้ไม่สามารถเข้าถึงข้อมูลโดยการ index ไม่สามารถมีข้อมูลซ้ำกันได้ และทำงานเร็วกว่า list

j = frozenset({"apple", "banana", "cherry"})     # * Data Type	frozenset ไม่เปลี่ยนรูปและแฮชได้ – เนื้อหาไม่สามารถเปลี่ยนแปลงได้หลังจากสร้างขึ้นแล้ว ดังนั้นจึงสามารถใช้เป็นคีย์พจนานุกรมหรือเป็นองค์ประกอบของชุดอื่นได้

k = True                                                       # ***Data Type	bool ค่า จริง ,เท็จ

l = b"Hello"                                                   # ** Data Type	bytes แปลงเป็น bytes จะเก็บอยู่ในรูปแบบของ binary.

n = bytearray(5)                                            # * Data Type bytearray แปลงวัตถุเป็นวัตถุ bytearray หรือสร้างวัตถุ bytearray ว่างตามขนาดที่ระบุ

m = memoryview(bytes(5))                            # * Data Type memoryview เป็นข้อมูลที่เก็บที่อยุ่ address ใน memory

o = None                                                      # *** Data Type NoneType เป็นค่าว่าง

print("a = ", a , "-"*5 ,"is of type ", "-"*5 , type(a)) #Hello World ----- is of type  ----- <class 'str'>

print("b = ", b,"-"*5 , "is of type ", "-"*5 , type(b)) #20 ----- is of type  ----- <class 'int'>

print("c = ", c ,"-"*5 , "is of type ","-"*5 , type(c)) #20.5 ----- is of type  ----- <class 'float'>

print("d = ", d ,"-"*5 , "is of type ", "-"*5 ,type(d)) # 1j ----- is of type  ----- <class 'complex'>

print("e = ", e , "-"*5 ,"is of type ", "-"*5 ,type(e)) #['apple', 'banana', 'cherry'] ----- is of type  ----- <class 'list'>

print("f = ", f ,"-"*5 , "is of type ", "-"*5 ,type(f)) #('apple', 'banana', 'cherry') ----- is of type  ----- <class 'tuple'>

print("g = ", g ,"-"*5 , "is of type ", "-"*5 ,type(g)) #range(0, 6) ----- is of type  ----- <class 'range'>

print("h = ", h , "-"*5 ,"is of type ","-"*5 , type(h)) #{'name': 'John', 'age': 36} ----- is of type  ----- <class 'dict'>

print("i = ", i , "-"*5 ,"is of type ","-"*5 , type(i)) #{'apple', 'banana', 'cherry'} ----- is of type  ----- <class 'set'>

print("j = ", j , "-"*5 ,"is of type ","-"*5 , type(j)) # frozenset({'apple', 'banana', 'cherry'}) ----- is of type  ----- <class 'frozenset'>

print("k = ", k , "-"*5 ,"is of type ", "-"*5 ,type(k)) # True ----- is of type  ----- <class 'bool'>

print("l = ", l , "-"*5 ,"is of type ","-"*5 , type(l)) #b'Hello' ----- is of type  ----- <class 'bytes'>

print("n = ", n , "-"*5 ,"is of type ", "-"*5 ,type(n)) #bytearray(b'\x00\x00\x00\x00\x00') ----- is of type  ----- <class 'bytearray'>

print("m = ", m , "-"*5 ,"is of type ","-"*5 , type(m)) #<memory at 0x7c8c49c0aa40> ----- is of type  ----- <class 'memoryview'>

print("o = ", o , "-"*5 ,"is of type ", "-"*5 ,type(o))  # None ----- is of type  ----- <class 'NoneType'>
```

Python - Modify Strings

---



Upper Case

```python
# Upper Case
a = "Hello, World!"
print(a.upper()) #HELLO, WORLD!
# ------------------------------------------------------------

# Lower Case
a = "Hello, World!"
print(a.lower()) #hello, world!

# Remove Whitespace
a = " Hello, World! "
print(a.strip()) # returns "Hello, World!"

# Replace String
a = "Hello, World!"
print(a.replace("H", "J")) #Jello, World!

# Split String
a = "Hello, World!"
print(a.split(",")) # returns ['Hello', ' World!']
```

Python - Escape Characters

---




Escape Characters


1.  \'    Single Quote
2.   \\  Backslash
1.   \n   New Line
2.   \r  Carriage Return
1.   \t   Tab
2.  \b   Backspace
1.   \f   Form Feed
2.   \ooo  Octal value
1.  \xhh   Hex value














```python
# txt = "We are the so-called "Vikings" from the north." #Error

txt = "We are the so-called \"Vikings\" from the north."

# Single Quote
txt = 'It\'s alright.'
print("txt : ", txt) #txt :  It's alright.

# Backslash
txt = "This will insert one \\ (backslash)."
print("txt : ", txt) #txt :  This will insert one \ (backslash).

# New Line
txt = "Hello\nWorld!"
print("txt : ", txt) #Hello
                        # World!

# Carriage Return
txt = "Hello\rWorld!"
print("txt : ", txt)

# Tab
txt = "Hello\tWorld!"
print("txt : ", txt) #txt :  Hello	World!

# Backspace
txt = "Hello \bWorld!"
print("txt : ", txt) #txt :  HelloWorld!

#	Octal value
txt = "\110\145\154\154\157"
print("txt : ", txt) #txt :  Hello

#	Hex value
txt ="\x48\x65\x6c\x6c\x6f"
print("txt : ", txt) #txt :  Hello
```
